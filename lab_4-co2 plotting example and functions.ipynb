```python
def read_and_parse_df():
    df = sm.datasets.co2.load_pandas().data
    df.reset_index(inplace = True)
    
    df.rename(columns={'index':'date'}, inplace = True)
    df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d').map(lambda d: d.year)
    
    use_df = df.groupby(df['date']).mean().reset_index()
    
    return use_df


check = read_and_parse_df()
```

```python
check.head()
```

```python
def plot_co2(use_df):
    fig, ax = plt.subplots(figsize= (15,7))
    ax.plot(use_df['date'], use_df['co2'], 'black')
    ax.set_title('Atmospheric CO2 Concentratin Increasing Over Time')
    ax.set_ylabel('CO2 Concentration (ppm)')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    ax.axhline(y = 350)
    ax.annotate('Such and Such Target', xy=(1958, 351), xycoords = 'data')
    #cite: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.axhline.html
    #cite: https://matplotlib.org/3.1.0/gallery/text_labels_and_annotations/annotation_demo.html
    plt.show() 

plot_co2(check)   
```

```python
def plot_co2(use_df, target_ppm, target_name):
    '''input is the cleaned and reformatted co2 df,
    an integer for the target pmm of cot
    a string for the name of that target
    output: displayes the plot
    '''
    fig, ax = plt.subplots(figsize= (15,7))
    ax.plot(use_df['date'], use_df['co2'], 'black')
    ax.set_title('Atmospheric CO2 Concentratin Increasing Over Time')
    ax.set_ylabel('CO2 Concentration (ppm)')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    ax.axhline(y = target_ppm)
    ax.annotate(target_name, xy=(1958, target_ppm+1), xycoords = 'data')
    #cite: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.axhline.html
    #cite: https://matplotlib.org/3.1.0/gallery/text_labels_and_annotations/annotation_demo.html
    plt.show() 

plot_co2(check, 340, 'Other Target') 
```

```python
def read_and_parse_df(period):
    df = sm.datasets.co2.load_pandas().data
    df.reset_index(inplace = True)
     
    df['year'] = pd.to_datetime(df['index'], format='%Y-%m-%d').map(lambda d: d.year)
    df['month'] = pd.to_datetime(df['index'], format='%Y-%m-%d').map(lambda d: d.month)
    # repetative, consider re-working
    
    use_df = df.groupby(df[period]).mean().reset_index()
    
    return use_df


co2_annually = read_and_parse_df('year')
```

```python
co2_annually.head()
```

```python
co2_monthly = read_and_parse_df('month')
co2_monthly.head()
```

```python
def plot_co2(use_df, xvalue, title, target_ppm = None, target_name = ''):
    '''input is the cleaned and reformatted co2 df,
    a string for xval indicating the column name used for x values ('month' or 'year')
    an integer for the target pmm of cot
    a string for the name of that target
    output: displayes the plot
    '''
    fig, ax = plt.subplots(figsize= (15,7))
    ax.plot(use_df[xvalue], use_df['co2'], 'black')
    ax.set_title(title)
    ax.set_ylabel('CO2 Concentration (ppm)')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    if target_ppm:
        ax.axhline(y = target_ppm)
        ax.annotate(target_name, xy=(1958, target_ppm+1), xycoords = 'data')
    #cite: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.axhline.html
    #cite: https://matplotlib.org/3.1.0/gallery/text_labels_and_annotations/annotation_demo.html
    plt.show() 

plot_co2(co2_annually, 'year','Atmospheric CO2 Concentratin Increasing Over Time', 
         340, 'Other Target')
```

```python
plot_co2(co2_monthly, 'month', 'Average CO2 Concentration by Month')
```
